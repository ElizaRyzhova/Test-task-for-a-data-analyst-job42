## 

Представлен список чисел. Необходимо вывести элементы исходного списка,
значения которых больше предыдущего элемента.

Подсказка: элементы, удовлетворяющие условию, оформить в виде списка.

Пример исходного списка: [300, 2, 12, 44, 1, 1, 4, 10, 7, 1, 78, 123, 55].\
Результат: [12, 44, 4, 10, 78, 123].

Реализуйте вариант без и с генераторным выражением

```python
# Решение без генераторного выражения
def find_elements(lst):
    result = []
    for i in range(1, len(lst)):
        if lst[i] > lst[i-1]:
            result.append(lst[i])
    return result

original_list = [300, 2, 12, 44, 1, 1, 4, 10, 7, 1, 78, 123, 55]
result = find_elements(original_list)
print(result)

# Решение с генераторным выражением
def find_elements(lst):
    return [lst[i] for i in range(1, len(lst)) if lst[i] > lst[i-1]]

original_list = [300, 2, 12, 44, 1, 1, 4, 10, 7, 1, 78, 123, 55]
result = find_elements(original_list)
print(result)

# Решение без применения циклов.
def find_elements(lst):
    return list(map(lambda x: x[1], filter(lambda x: x[1] > x[0], zip(lst, lst[1:]))))

original_list = [300, 2, 12, 44, 1, 1, 4, 10, 7, 1, 78, 123, 55]
result = find_elements(original_list)
print(result)
```

## 
Преобразовать слова «разработка», «администрирование», «protocol»,
«standard» из строкового представления в байтовое и выполнить
обратное преобразование (используя методы encode и decode).

```python
words = ["разработка", "администрирование", "protocol", "standard"]

# Преобразование строкового представления в байтовое
encoded_words = [word.encode('utf-8') for word in words]
print("Закодированные слова:", encoded_words)

# Обратное преобразование байтового представления в строковое
decoded_words = [word.decode('utf-8') for word in encoded_words]
print("Декодированные слова:", decoded_words)
```

## 
Написать код выгрузки курса сенегальского (западноафриканского) франка XOF-CFA Frank в рублях на дату с 01.01.2023 по настоящее время.
В результате должна получиться таблица с колонками: 
- Дата
- Валюта
- Курс

```python
from datetime import date, timedelta
import requests
import csv

def fetch_currency_rates(currency_code):
    # Установка начальной и конечной даты
    start_date = date(2023, 1, 1)
    end_date = date.today()

    # Определение колонок таблицы
    columns = ['Дата', 'Валюта', 'Курс']
        
    # Вывод заголовка таблицы
    print('{:<12}{:<}{:<10}'.format(columns[0], columns[1], columns[2]))

    # Создание пустого списка для хранения данных
    data = []

    # Получение курса выбранной валюты для каждой даты
    current_date = start_date
    while current_date <= end_date:
        url = f'https://www.cbr-xml-daily.ru/archive/{current_date.strftime("%Y/%m/%d")}/daily_json.js'
        response = requests.get(url)
        if response.status_code == 200:
            json_data = response.json()
            valutes = json_data['Valute']

            
            if currency_code in valutes:
                selected_valute = valutes[currency_code]
                data.append([current_date.strftime('%d.%m.%Y'), currency_code, selected_valute['Value']])
                print('{:<12}{:<}{:<10}'.format(current_date.strftime('%d.%m.%Y'),selected_valute['CharCode'], selected_valute['Value']))
            else:
                available_currencies = valutes.keys()
                print("Доступные валюты:", ", ".join(available_currencies))
                print("Выбранный код валюты не найден.")
                break
        current_date += timedelta(days=1)

    # Сохранение данных в CSV-файл
    filename = f'{currency_code}_rates.csv'
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(columns)
        writer.writerows(data)

    print(f"Таблица успешно сохранена в файл: {filename}")
```

```python
# Получение от пользователя кода выбранной валюты
# currency_code = input("Введите код выбранной валюты (например, EUR или XOF): ")

currency_code = 'CFA'
fetch_currency_rates(currency_code.upper())
```
